```python
def swaped(arr, i, j):
    """
    docstring:function to swaping two different position of an array
    input:
        arr: array
        i: position number one
        j: position number two
    output:
        return the array already swaped
    """
    temp = arr[i]
    arr[i] = arr[j]
    arr[j] = temp

def partition(arr,low,high, column, ascending = True):
    """
    Docstring: function to use for pivoting in the quick sort
    input:
        arr: array for partitioning
        low: lowest position
        high: highest position
        column: column use for sorting
    """
    if ascending:
        #get the value for pivoting
        pivot = arr[high][column]
        i = low - 1
        #pivoting
        for j in range(low, high):
            if arr[j][column] < pivot:
                i+=1
                swaped(arr,i,j)

        swaped(arr,i+1,high)
        return i+1  
    else:
        #get the value for pivoting
        pivot = arr[high][column]
        i = low - 1
        #pivoting
        for j in range(low, high):
            if arr[j][column] > pivot:
                i+=1
                swaped(arr,i,j)

        swaped(arr,i+1,high)
        return i+1  

def quick_sort(arr, low, high, column, ascending = True):
    """
    Docstring: function for quick sorting
    input:
        arr: array for sorting
        low: lowest position
        high: highest position
        column: column use for sorting
    output:
        return a sorted array and time of execution
    """
    t1 = time()
    if low < high:    
        """ pi is partitioning index, arr[p] is now  
        at right place """
        pi = partition(arr, low, high, column, ascending)
  
        #Separately sort elements before  
        #partition and after partition  
        quick_sort(arr, low, pi - 1, column, ascending)  
        quick_sort(arr, pi + 1, high, column, ascending)  
    t2 = time()
    return [arr,t2-t1]
```

```python
def seq_search(array, dato, columna):
    """
    Docstring
    """
    t1 = time()
    n = len(array)
    encontrado = []
    """
    array, encontrado are list
    n and dato are int
    """
    j = 0
    for i in range(n):
        if dato == array[i][columna]:
            encontrado.append(i)
            #elementos encontrados
            j=1
        elif j == 1:
            break
    if len(encontrado) == 0:
        return [-1,time()-t1]
    t2 = time()
    print("searching time: ", t2-t1)
    return [encontrado, t2-t1]
```

```python
def sorting(dataframe, column, ascending = True):
    """
    Docstring: function to sort a dataframe
    input:
        dataframe: it is the table or dataframe we wanna sort
        column: the column name we want the dataframe be sorted by
    output:
        sorted dataframe
    """
    #getting the columns' names of the dataframe
    data_columns = dataframe.columns.tolist()
    #getting the position of the column
    colum_index = data_columns.index(column)
    #getting the dataframe into list of lists
    data = dataframe.values.tolist()
    #getting the lenght of the data, i.e- number of registers
    n = len(data)
    #sorting the data
    datos = quick_sort(data,0,n-1,colum_index, ascending)
    #transforming the sorted data into a dataframe
    data_final = pd.DataFrame(datos[0], columns = data_columns)
    print("sorting time: ",datos[1])
    return data_final
```

```python
def searching(DataFrame, column, dato):
    """
    Docstring: function for searching a value in the dataframe
    input:
        DataFrame: dataframe
        column: column name where the value is
        dato: value we are looking for
    """
    #sorting the data, hence We can use sequential search
    dataframe = sorting(DataFrame, column)
    #getting the columns' names of the dataframe
    data_columns = dataframe.columns.tolist()
    #getting the position of the column
    colum_index = data_columns.index(column)
    #getting the dataframe into list of lists
    data = dataframe.values.tolist()
    n = len(data)
    #getting the lenght of the data, i.e- number of registers
    datos = seq_search(data, dato, colum_index)
    if datos[0] == -1:
        return DataFrame
    #transforming the sorted data into a dataframe
    data_final = pd.DataFrame(data[datos[0][0]:datos[0][len(datos[0])-1]+1], columns = data_columns)
    return data_final
```

```python
def multiple_searching(DataFrame, columnas, datos):
    """
    Docstring: function for multiple searching values on the dataframe
    input:
        DataFrame: dataframe
        columnas: columnas where we are searching values
        datos: datos we are searching
    """
    t1 = time()
    #making a copy of the dataframe, this is only for security
    searched_data = DataFrame.copy()
    #making the index a column
    searched_data = searched_data.reset_index(drop = False)
    #for to iterate in searching for each column and its value
    for i in range(len(columnas)):
        try:
            #searching individuality
            searched_data = searching(searched_data, columnas[i], datos[i])
        except:
            print("{} Does not exist on column {}".format(datos[i],columnas[0]))
    
    """index_for_elimination = searched_data['index'].tolist()
    searched_data.drop('index',axis = 1, inplace = True)"""
    #returning data
    print("Total time: ",time()-t1)
    return searched_data
   #, index_for_elimination
        
```

```python
def validar(DataFrame, indices):
    """
    Docstring: function that check if the indices given are valid
    input:
        DataFrame: dataframe
        indices: Indices of the values to eliminates
    output:
        return the indices that are valid
    """
    validos = []
    #getting the dataframe into list of lists
    data = DataFrame['index'].tolist()
    for i in indices:
        try:
            if int(i) in data:
                validos.append(int(i))
            else:
                print("{} is not valid".format(i))
        except:
            print("{} is not valid".format(i))
    return validos
```

```python
def elimination_function(original_dataframe,DataFrame, indices = None, todos = False):
    """
    Docstring: function that check if the indices given are valid
    input:
        original_dataframe: original dataframe where you will eliminate the data
        DataFrame: dataframe
        indices: Indices of the values to eliminates
        tods: when you wanna eliminate all the elements
    output:
        return the new dataframe without the eliminated values
    """
    if todos:
        indices_eliminar = DataFrame['index'].tolist()
        new_dataframe = original_dataframe.drop(indices_eliminar, axis = 0)
        return new_dataframe
    else:
        validados = validar(DataFrame, indices)
        new_dataframe = original_dataframe.drop(validados, axis = 0)
        return new_dataframe
```
