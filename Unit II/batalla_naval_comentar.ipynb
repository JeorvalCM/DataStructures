```python
#Here we create a function that will return a list of just one dimention, this is for us to be easier to handle
def flattenerclass(lista):
    try:
        #return a list of one dimension
        flattenerclass([principal for sub in lista for principal in sub])
    except:
        return [principal for sub in lista for principal in sub]
```

```python
class queue:
    #initializer
    def __init__(self, item = None):
        try:
            self.items = flattenerclass(item)
        except:
            self.items = item
            
        if self.items == None:
            self.items = []
            
    #check if the stack is empty
    def empty(self):
        return self.items == []
    
    #adding a new element at the end
    def push(self, item):
        try:
            if(len(item)) >0:
                for i in item:
                    self.items.append(i)
                return
        except:
            return self.items.append(item)
    
    #eliminating the last itam
    def pop(self):
        return self.items.pop(0)
    
    #check the stackÂ´s size
    def size(self):
        return len(self.items)
    
    #print the content of the stack
    def content(self):
        return self.items
```

```python
class NodeTree:
    listapre = []
    listapos = []
    listain = []
    #initializer of the class
    def __init__(self,value = None, datos = None):
        #left son
        self.left = None
        #right son
        self.right = None
        #data save in the node
        self.valor = value
        #data saved
        self.data = datos
        
    
    #function to add a new element
    def add_element(self, value, datos):
        #checking if the number is fewer than the number of the father node
        if value <= self.valor:
            #adding the data to the left son
            self.add_left(value, datos)
        else:
            #adding the data to the right son
            self.add_right(value, datos)
    
    #function to add a node in the left son
    def add_left(self, value, datos):
        #checking if its empty the left son
        if self.left == None:
            #adding a new node in the left son
            self.left = NodeTree(value, datos)
        else:
            #going into recursion until find an empty son
            self.left.add_element(value,datos)
    
    #function to add a node in the right son  
    def add_right(self,value, datos):
         #checking if its empty the right son
        if self.right == None:
            #adding a new node in the left son
            self.right = NodeTree(value, datos)
        else:
            #going into recursion until find an empty son
            self.right.add_element(value, datos)
   
                        
    #preorder algorithm
    def preorder(self, lista):
        #print the value of the node or rather its number
        lista.append(self.valor)
        print("Nodo value ", self.valor)
        #check if there is a node in the left son
        if self.left != None:
            #recursive function
            self.left.preorder(lista)
        #check if there is a node in the right son
        if self.right != None:
            #recursive function
            self.right.preorder(lista)    
        else:
            #case where the node does not have sons
            return
        
    #inorder algorithm
    def inorder(self, lista):
        #check if there is a node in the left son
        if self.left != None:
            #recursive function
            self.left.inorder(lista)
        lista.append([self.valor, self.data])
        #print the value of the node or rather its number
        print("Nodo value ", self.valor)
        #check if there is a node in the right son
        if self.right != None:
            #recursive function
            self.right.inorder(lista)    
        else:
            #case where the node does not have sons
            return
        
    #posorder algorithm    
    def posorder(self, lista):
        #check if there is a node in the left son
        if self.left != None:
            #recursive function
            self.left.posorder(lista)
        #check if there is a node in the right son
        if self.right != None:
            #recursive function
            self.right.posorder(lista)
        lista.append(self.valor)
        #print the value of the node or rather its number
        print("Nodo value ", self.valor)
         #case where the node does not have sons
        return
    
     
    #function to eliminate nodes
    def delete(self, number):
        #check if the value is the same as the one we want to eliminate
        if self.valor == number:
            #checking if it has children
            if self.hasChildren():
                #checkinf if it has two children
                if self.hasbothchildren(): 
                    if self.left.valor <= self.right.valor:
                        self.valor = self.left.valor
                        self.left = None
                    else:
                        self.valor = self.right.valor
                        self.right = None
                
                #checking if the son is in the left
                elif self.hasleftson():
                    self.valor = self.left.valor
                    self.right = self.left.right
                    self.left = self.left.left
                
                #checking if the son is in the left
                elif self.hasrightson():
                    self.valor = self.right.valor
                    self.right = self.right.right
                    self.left = self.right.left
                    
            #elimination when it does not have children
            else:
                self.valor = None
        else: 
            #check if there is a node in the left son
            if self.left != None:
                #recursive function
                self.left.delete(number)
            #check if there is a node in the right son
            if self.right != None:
                #recursive function
                self.right.delete(number)    
            else:
                #case where the node does not have sons
                return

                    
    
        
    
    #function for checking if it has son
    def hasChildren(self):
        return self.left != None or self.right != None
    
    #function to check if has two children
    def hasbothchildren(self):
        return self.left != None and self.right != None
    
    #function for checking it has a left son
    def hasleftson(self):
        return self.left != None
    
    #function for checking it has a right son
    def hasrightson(self):
        return self.right != None
    
    #functions to get the lists of the search algorithm
    def preorder_list(self):
        lista = []
        self.preorder(lista)
        return lista
    #Here we return a list with the nodes in inoderder
    def inorder_list(self):
        lista = []
        self.inorder(lista)
        return lista
    #Here we return a list with the nodes in posorder 
    def posorder_list(self):
        lista = []
        self.posorder(lista)
        return lista
```
